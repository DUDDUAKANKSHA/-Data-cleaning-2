```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.datasets import load_iris
data=load_iris()
x = pd.DataFrame(data.data,columns=data.feature_names)
y = data.target
print("Original Features:\n",x.head())
k = 2
selector = SelectKBest(score_func=f_classif,k=k)
x_new=selector.fit_transform(x,y)
print("Transformed Features:\n",x_new)
feature_scores=selector.scores_
print("Scores of Features:\n",feature_scores)
selected_features = x.columns[selector.get_support ()]
x_selected=pd.DataFrame(x_new,columns=selected_features)
print("Selected Features:\n",selected_features)
print("Selected Features columns:\n",x_selected.head())
print("/nFeature Scores:")
for feature,score in zip(x.columns,feature_scores):
  print(f"{feature}: {score:.2f}")
#visulaization of feature importance
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.barh(x.columns,feature_scores,color='skyblue')
plt.axvline(x=feature_scores[selector.get_support()].min(),color='red',linestyle='--',label='Selection Threshold')
plt.xlabel('Score')
plt.ylabel("Feature")
plt.title('Feature Score (SelectedKBest)')
plt.legend()
plt.show()

```
